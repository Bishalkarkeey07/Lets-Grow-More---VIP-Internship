**Build the Models**

```python

X_train = df.drop("class", axis=1)
Y_train = df["class"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

# splitting data in training set(80%) and test set(20%).
x_train, x_test, y_train, y_test = train_test_split(X_train, Y_train, test_size=0.2)
```

**Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt=dt.fit(x_train, y_train)


act = accuracy_score(y_train,dt.predict(x_train))
print('Training Accuracy is: ',(act*100))

act = accuracy_score(y_test,dt.predict(x_test))
print('Test Accuracy is: ',(act*100))
```

**KNN CLASSIFICATION**

```python

from sklearn.neighbors import KNeighborsClassifier     
knn = KNeighborsClassifier()
knn = knn.fit(x_train, y_train)

act = accuracy_score(y_train,knn.predict(x_train))
print('Training Accuracy is: ',(act*100))

act = accuracy_score(y_test,knn.predict(x_test))
print('Test Accuracy is: ',(act*100))
```

**Random Forest Classifier**

```python

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf = rf.fit(x_train,y_train)

act = accuracy_score(y_train,rf.predict(x_train))
print('Training Accuracy is: ',(act*100))

act = accuracy_score(y_test,rf.predict(x_test))
print('Test Accuracy is: ',(act*100))
```

**CONCLUSION : If we use Decision Tree Classifier for iris flower classification then the test accuracy is 96.67%, if KNN is used then the test accuracy is 100% ,if Random forest classifier is used then the test accuracy is also 96.67%.**
